```python
from bs4 import BeautifulSoup
import requests
```

## Retriving Data

```python
web = requests.get('https://en.wikipedia.org/wiki/Wikipedia')
print(web)
```

```python
# web.content -- used to load the entire code and data of a website and you can use a get function to get any specific data or function
```

```python
web.url # load the url of a website you have loaded
```

```python
web.status_code # load the status code 
```

## Parsing code 

```python
# We created a Soup object that allows us to navigate and search through the HTML data efficiently using BeautifulSoup’s methods.
soup = BeautifulSoup(web.content,"html.parser") 
```

```python
# Using Prettify we have listed the code in a structured format 
print(soup.prettify())
```

```python
soup.title # Accesing the title using soup methods
```

```python
soup.title.name
```

```python
soup.p # Accessing the first paragraph tag
```

```python
soup.a # Accessing the first anchor tag
```

```python
soup.h1 # Accessing the first title tag
```
